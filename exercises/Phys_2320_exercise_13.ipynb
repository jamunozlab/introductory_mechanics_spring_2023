The word 'torque' is often represented with the Greek letter $\tau$ (tau) and has a very concrete definition in physics (and all the sciences and engineerings use the same definition). The definition is:

$\tau = \vec{r} \times \vec{F} \;$,

where $\vec{r}$ is a position vector that goes from the point about where the torque is being measured (the axis of rotation) to the point where the force is being applied, and $\vec{F}$ is the force vector.

<hr>


The cross on the right-hand side of the equation above represents a mathematical operation called the $vector$ $product$ (it is often also referred to as the cross product). The operation is between two vectors and, as the name implies, the result is another vector. 

.

Just like vector addition and multiplication of a vector by a scalar, and like the dot product, the cross product has both a geometric and an algebraic definition. It is the following for vectors $\vec{A}$ and $\vec{B}$:

$ \vec{A} \times \vec{B} = \hat{n} |\vec{A}||\vec{B}|\sin{\theta_{AB}} $ $\;$ $\;$ (geometric)

$ \vec{A} \times \vec{B} = \hat{i} (A_yB_z-A_zB_y) - \hat{j} (A_xB_z-A_zB_x) + \hat{k} (A_xB_y-A_yB_x)$ $\;$ $\;$ (algebraic)

.

In the geometric definition, the magnitude of $\vec{A}$ (denoted $|\vec{A}|$) is multiplied by the magnitude of $\vec{B}$ (denoted $|\vec{B}|$), and that number is then multiplied times $\sin{\theta_{AB}}$, the sine of the angle from $\vec{A}$ to $\vec{B}$ (in that direction!). 

.

Unlike the dot product, the cross product produces another vector, and vectors consists of magnitude and direction. The direction of the vector is given by $\hat{n}$, which is the unit vector (magnitude is 1) that is normal (at a 90 degree angle) to the plane created by vectors $\vec{A}$ and $\vec{B}$. The direction is given by the right-hand rule.

.

In the algebraic definition, the components of $\vec{A}$ and $\vec{B}$ along each direction (denoted $A_x$ and $B_x$ for the $x$-direction, $A_y$ and $B_y$ for the $y$-direction, $A_z$ and $B_z$ for the $z$-direction) are multiplied and then added together, but in a special fashion. Notice that all three dimensions are potentially represented, since $\hat{i}$, $\hat{j}$, $\hat{k}$ are the unit vectors along the $x-$, $y-$, and $z-$directions, respectively. The components of the vectors that are multiplied are always the ones that are orthogonal to the resulting direction. For example, in the first term on the right-hand side of the algebraic definition equation, $\hat{i} (A_yB_z-A_zB_y) $, which gives you the component of the resulting vector along the $x-$direction, you find the $y$ and $z$ components of the original vectors, and so on.

.

 An easy way to remember the algebraic definition of the cross product is to remember that it is the determinant of the matrix given by:

$\vec{A} \times \vec{B} = 
\begin{vmatrix}
     \hat{i} & \hat{j} & \hat{k}\\ 
     A_x & A_y & A_z\\
     B_x & B_y & B_z 
\end{vmatrix}$
.
