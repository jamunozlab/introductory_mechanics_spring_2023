Consider a bowling ball with a diameter $d=22$ cm (so its radius $R = 11$ cm) and for now think about it as a solid sphere of uniform mass density $\rho$. Its total mass is $M=5$ kg and the magnitude of its angular momentum is $|\vec{L}|=0.23$ kg$\cdot$m$^2$/s. You are asked to calculate at what angular velocity $\vec{\omega}$ it must be rotating. A table of moments of intertia $I$ for several common shapes and axes of rotation calculated analytically is given below.

phpRvs8d4.png

Since the angular momentum $\vec{L}$ is given by:

.

$\vec{L} = I\vec{\omega} = \vec{r} \times \vec{p} \,$ ,

.

and the moment of inertia of a solid sphere of uniform density is $I = \frac{2}{5}MR^2 \,$ , 

.

$|\vec{\omega}| = \frac{|\vec{L}|}{I} = \frac{|\vec{L}|}{\frac{2}{5}MR^2} = \frac{5 |\vec{L}|}{2 MR^2} \,$ .

.

Boringly plugging in some numbers we get, $|\vec{\omega}| = \frac{5(0.23 {\rm kg}\cdot{\rm m}^2/{\rm s})}{2 (5 {\rm kg})(0.11 {\rm m})^2} = 9.5 {\rm /s}$

.

Since there are $2\pi$ radians in a revolution and 60 seconds in 1 minute,  $|\vec{\omega}| = \frac{9.5 {\rm rad}}{{\rm s}} \times \frac{1 {\rm\, revolution}}{2\pi \,{\rm rad}} \times \frac{60 {\rm s}}{1 {\rm min}} = 90.8 $ rpm


<hr>

Nevertheless, being quite the skeptic, you don't trust calculus. You will calculate the value of the moment of inertia on your own, call it $I_{\rm own}$. From the definition of the moment of inertia,

.

$I_{\rm own} = \sum_i m_i r_i^2 \,$ .

You start by creating a cube made of smaller cubes. Each of the smaller cubes is a *voxel*, which is basically a 3 dimensional pixel and will serve as your basic particle $i$. Remember that $r_i$ is the "parallel" distance from the location of particle $i$ to the axis of rotation. This is the shortest possible distance to any point on the axis. Since you are using the particle model (although now with a bunch of particles), all mass is located at a single point, which will be the center of the voxel. Assume, without loss of generality, that the bowling ball is rotating about the z-axis.

````python
# make our life easy
import numpy as np
````

````python
# The following creates a series of 21 points from negative r = -0.11 m to positive r = +0.11 m
# You get a series of points along a line with length 0.22 m (the diameter of the ball)

num_points = 11
radius = 0.11 #m
mass = 5 #kg
np.linspace(start=-radius, stop=radius, num=num_points, endpoint=True)
````

````python
# the following loop gives you the points that lie within the sphere of radius r = 11 m

x_ = np.linspace(start=-radius, stop=radius, num=num_points, endpoint=True)
y_ = np.linspace(start=-radius, stop=radius, num=num_points, endpoint=True)
z_ = np.linspace(start=-radius, stop=radius, num=num_points, endpoint=True)

sphere_grid = []

for x in x_:
  for y in y_:
    for z in z_:
      if np.sqrt(x**2 + y**2 + z**2) <= radius:
        sphere_grid.append((x, y, z))
````

````python
# The resulting grid looks like the one below
# Some grid points lie within the sphere (we selected them)
# Some don't (we discarded them)
````

````python
# Statistics on the resulting grid
print("Total number of grid points:", num_points**3)
print("Total number of grid points inside the sphere:", len(sphere_grid))
````

````python
# Distribute the total mass M = 5 kg uniformily among all the grid points inside the sphere
mass_per_particle = mass / len(sphere_grid)
print("Mass per particle:", mass_per_particle, "kg")
````

````python
def compute_moment_of_inertia_sphere(sphere_grid, mass_per_particle):
  # And magic happens
  I_own = 0
  for x, y, z in sphere_grid:
    I_own = I_own + mass_per_particle * (x**2 + y**2)

  return I_own
````

````python
# Show the result of your proud work 
moment_of_inertia = compute_moment_of_inertia_sphere(sphere_grid, mass_per_particle)
print("The moment of inertia computed numerically is:", moment_of_inertia, "Kg m^2")
````
