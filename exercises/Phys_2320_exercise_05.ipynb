Humans come up with all sorts of contorted ways to make the addition of vectorial forces, in our minds, easier. "Take the sine of the angle between the negative part of the vertical axis and the force vector in the clockwise direction!" one shouts. "No! Take the cosine of the angle between the negative section of the horizontal axis and the force vector in the counter-clockwise direction!" another angrily replies. In reality, this all obscures the simplicity of physics. The laws of physics are the same but independent in orthogonal directions. If we, in our desire to compute one less hypothenuse, rotate the Cartesian axes to align them with as many forces as possible, nature could not care less. 

Consider the situation portrayed below, in which 4 forces of different magnitude and orientation are applied to a poor particle. The axes were rotated in such a way that they are aligned with 3 of those forces. This way, only the 3.0 N force has components in both $x$ and $y$. The sum of forces is then

.

$ \sum F_x = -1.0{\rm N} - 3.0{\rm N} \sin \theta + 5.0{\rm N}$

$ \sum F_y = - 3.0{\rm N} \cos \theta + 2.82{\rm N}$

.

If the angle $\theta = 20^{\rm o}$, as in the homework problem, then:

$ \sum F_x = -1.0{\rm N} - 3.0{\rm N} (0.34) + 5.0{\rm N} = 3.0 {\rm N}$

$ \sum F_y = - 3.0{\rm N} (0.94) + 2.82{\rm N} = 0.0 {\rm N}$

In the back of your mind, you have always hated such arbitraryness. You decide, quite reasonably if I might add, that you will always just consider the angle in the counter-clockwise (positive) direction between the (positive) horizontal axis and each vector, and you will forget about everything else. Let them call you a right-hand chauvinist, you know they are darn wrong: you could just as easily have chosen a left-hand Cartessian system. Consistency is what matters. Done this way, the right-hand way, the cosine of such angle will be associated with the horizontal axis, and the sine with the vertical axis. Thought often ommited because people are lazy even to write things down, the sines and cosines are always there. This is how the components of the net force will look like in such situation:

.

$ \sum F_x = 5.0{\rm N}\cos 0^{\rm o} + 2.82{\rm N}\cos 90^{\rm o} + 1.0{\rm N}\cos 180^{\rm o} + 3.0{\rm N}\cos 250^{\rm o} $

$ \sum F_y = 5.0{\rm N}\sin 0^{\rm o} +  2.82{\rm N} \sin 90^{\rm o} + 1.0{\rm N}\sin 180^{\rm o} + 3.0{\rm N} \sin 250^{\rm o} $

.

Since $\cos 0^{\rm o} = 1; \cos 90^{\rm o} = 0; \cos 180^{\rm o} = -1; {\rm and} \cos 250^{\rm o} = -0.34;$

Also $\sin 0^{\rm o} = 0; \sin 90^{\rm o} = 1; \sin 180^{\rm o} = 0; {\rm and} \sin 250^{\rm o} = -0.94;$

.

You suddenly realize that there is reason behind this madness! If only you could write some code to solve this class of problems in general...

```python
import numpy as np
```

```python
def net_force_x(vectors): # The function argument named "vectors" is a list of grouped values known as "TUPLES", that represent magnitude and angle respectivley
  net_force = 0 # net force is set to 0 before function loop beigns to avoid computational errors
  for vector in vectors: # this for loop will iterate for Tuple held in the list "vectors"
    magnitude = vector[0] # the first value in the first tuple of our "vector" list represents the forces magnitude
    angle = vector[1] # The second value in the first tuple of out "vector" list represents the forces' angle in respect to the POSITIVE X-AXIS
    net_force = net_force + magnitude * np.cos(np.deg2rad(angle)) #This line calculates the net force values of each group of forces in respect to the x-axis
                                                                  # The net force emmitted by each group is then summed and output to the user
  return net_force
```

```python
def net_force_y(vectors): # much like the fuction above this code cell computes the net force, but this time in respect to the Y-axis
  net_force = 0           # Given the list variable named "vectors" 
  for vector in vectors:  # Remember that the tuples within "vectors" are a group of 2 values EX. (5,0)
    magnitude = vector[0] # in the case of the tuple (5,0) the value "5" represent a magnitude of 5N, while the value "0" respresents an angle of "sin(0)"
    angle = vector[1]
    net_force = net_force + magnitude * np.sin(np.deg2rad(angle))

  return net_force
  # The two code cells described above are simply the function definitions, farther down this page we will perform what is called a "Function call".
  # to utilize these computations
```

```python
# The functions above take a list of vectors
# Each vector is specified by its magnitude in Newtons
# and its angle with respect to the positive horizontal axis
# For the previous problem: 

vectors = [(5, 0), (2.82, 90), (3, 250), (1, 180)] # Each set of grouped values is known as a tuple, the first value in each set represent the magnitude of force
                                                   # While the second value in each group represents the angle in respect to the positive X-axis.

# This "vector" data will serve as the input to the two functions defined above that calculate the net force in the X and Y axes
```

```python
# The net force (in Newtons) along the horizontal axis is
net_force_x(vectors)

#This line of code is a "function call", it simply commands the computer to execute the functions that we already defined above 
```

```python
# The net force along the vertical axis is
net_force_y(vectors)

#This line of code is a "function call", it simply commands the computer to execute the functions that we already defined above 
```

```python
#In agreement with our previous calculations! 
# The magnitude of such force is, of course, 
print(np.sqrt(net_force_x(vectors)**2 + net_force_y(vectors)**2), 'Newtons') # sqrt[(fx)^2 + (fy)^2]

# This code cell solves for a net magnitude between the X and Y axes using the pythagorean thereom 
```

```python
# The angle is 
print(np.rad2deg(np.arctan(net_force_y(vectors) / net_force_x(vectors))), 'degrees') # arctan[(fy)/(fx)]=

#This code cell solves for the net angle (direction) of the force between the x and y axes, agaim using a formula deriived from pythagoreon theoreom. 
```
