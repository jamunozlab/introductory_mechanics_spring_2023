The word 'work' is often represented with the letter $W$ and has a very concrete definition in physics (and all the sciences and engineerings use the same definition). The definition is:

$W = \int_C \vec{F} \cdot d\vec{s} \;$,

where $\vec{F}$ is the force vector and $d\vec{s}$ is the derivative of the displacement vector. In the special case in which the force is constant and hence we can take it out of the integral, 

$W = \vec{F} \cdot \int_C   d\vec{s} = \vec{F} \cdot \vec{s}|_{s_0}^{s_f} = \vec{F} \cdot \Delta \vec{s} \;$,

which is a definition you might be familiar with.

<hr>


The little dot inside the integral and later propagated is important: it represents a mathematical operation called the $scalar$ $product$ (it is often also referred to as the dot product). The operation is between two vectors and, as the name implies, the result is a scalar. The $"C"$ next to the integral symbol means that the integral is along the whole path of $\vec{s}$.

.

Just like vector addition and multiplication of a vector by a scalar, the dot product has both a geometric and an algebraic definition. It is the following for vectors $\vec{A}$ and $\vec{B}$:

$ \vec{A} \cdot \vec{B} = |\vec{A}||\vec{B}|\cos{\theta_{AB}} $ $\;$ $\;$ (geometric)

$ \vec{A} \cdot \vec{B} = A_xB_x + A_yB_y + A_zB_z + ... $ $\;$ $\;$ (algebraic)

.

In the geometric definition, the magnitude of $\vec{A}$ (denoted $|\vec{A}|$) is multiplied by the magnitude of $\vec{B}$ (denoted $|\vec{B}|$), and that number is then multiplied times $\cos{\theta_{AB}}$, the cosine of the angle from $\vec{A}$ to $\vec{B}$ (in that direction!).

.

In the algebraic definition, the components of $\vec{A}$ and $\vec{B}$ along each direction (denoted $A_x$ and $B_x$ for the $x$-direction, $A_y$ and $B_y$ for the $y$-direction, and so on) are multiplied and then added together. Notice that your vectors can have 2, 3 or even more dimensions.
