In a previous exercise, you learned how to use the following functions to obtain the components of a net force along the horizontal axis (net_force_x) and along the vertical axis (net_force_y). Evaluate the following cells.



```python
import numpy as np
```

```python
# Just like the code blocks seen in Exercise #5, this function receives a list of paired of magnitudes and angles named "Vectors"
# It then multiplies the magnitude by the Cos(Î¸) [The angle specified in our vector tuple] to calculate the componet of the specific componet of the specific force vector
# Finally, It sums  the Force componets, and stores them in the variable named "net_force"

def net_force_x(vectors): 
  net_force = 0
  for vector in vectors:
    magnitude = vector[0]
    angle = vector[1]
    net_force = net_force + magnitude * np.cos(np.deg2rad(angle))

  return net_force

  # We will continually call this function during this workshop, 
  # it's usage is essential to the computations performed here, but the governing concepts and formulas coincide with the laws of motion exactly.
```

Consider the situation below in which Kojiro 'Steve' Hyuga is training for the upcoming FIFA World Cup by pushing a crate. There are 4 vectors involded: 

1) The weight $\vec{w}$ of the crate (mass $\times$ acceleration due to gravity)

2) The normal force $\vec{N}$ produced by the floor that does not want to let the crate fall through (equal in magnitude to 'weight', and always perpendicular to the surface)

3) The push $\vec{p}$ produced by Steve's ambitious muscles
 
4) The force of friction $\vec{f}$ which opposes movement (magnitude is $\mu |\vec{N}|$ and direction is opposite of $\vec{p}$).

How can we find the force Steve needs to exert for the crate to start moving if the crate has a rubber bottom and the surface is dry concrete?

```python
# Here we define some constant variables that will be referenced continually throughout this exercise, be sure to familiarize yourself with its contents

mass = 50 #kg   # The crate has a Mass of 50kg
g = 9.8 #m/s^2  # The Acceleration caused by Earth's gravitational field
w = mass * g    # Force = Mass * Acceleration, in our specific case here (Weight = mass * Acceleration caused by gravitational field)
N = w           #The Normal Force is equal in magnitude to the "weight" but opposite in direction
push = 491 # Newtons (Force exherted by Kojiro)
mu = 1.0 # Coefficient of Static friction (Obtained from the table above)
```

```python
# Here we define the vector variables that will hold the data for magnitude and angle of each of our 4 forces
# These will be used in conjunction with our function call shortly to caluculate the sum of forces

w_vector = (w, 270) # The magnitude of the force "Weight" points "strait down"
N_vector = (w, 90)  # The magnitude of the "Normal Force" points "strait up"
f_vector = (mu * N, 180) # The magnitude of the "Force of friction" points "to the Left" [cos(180â) = -1]
p_vector = (push, 0) # The magnitude of "Push" points "to the Right" [cos(0â) = 1]

vectors = [w_vector, N_vector, p_vector, f_vector] # This line of code simply assembles a list of our 4 vectors just previously defined above.
```

```python
print(net_force_x(vectors), 'Newtons') # Here we call our "net_force_x" function and provide our newly created "vectors" list as an argument. 
```

The result of our "net_force_x" function is a positive value, meaning of course that the force of static friction has just been overcome and the box has began to move.
As soon as the crate starts moving, kinetic friction kicks in. 

1. What would be the acceleration of the box after static frictions is broken?

```python
mu = 0.7 # Coefficient of Kinetic friction (Obtained from the table at the top of this page)
```

```python
# All of our "vector" values remain the same apart from the "Force friction"
w_vector = (w, 270)
N_vector = (w, 90)
f_vector = (mu * N, 180) # Since we have decrease the value of "mu" to correspond with Kinetic friction, this forces' magnitude will decrease once computed
p_vector = (push, 0)
vectors = [w_vector, N_vector, p_vector, f_vector]
```

```python
print(net_force_x(vectors), 'Newtons') # Function call "net_force_x"
```

```python
# Now we will calculate the Acceleration of the crate caused by the "net_force_x"
# Remember that (Force=Mass*Acceleration) thus, with a little rearrangement we have (Acceleration=Force/Mass)
         
acceleration = net_force_x(vectors) / mass 
print(acceleration, 'm/s^2') 
```

```python
# With our newfound knowledge of acceleration above, we can now approximate the final Velocity of the crate for any given Time
# Remember that (Acceleration=Velocity/ÎTime), rearranging this equation as well we can see that (Velocity=Acceleration*ÎTime)

velocity_initial = 0 #seconds
time = 5 #seconds
velocity_final = velocity_initial + acceleration * time
print(velocity_final, 'm/s') 
```
